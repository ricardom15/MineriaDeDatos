#Bollinger Band 
Las bandas de Bollinger son unos indicadores utilizados en el análisis técnico de los mercados financieros. Fueron introducidos por John Bollinger en los años 1980. La representación gráfica de las bandas de Bollinger son dos curvas que envuelven el gráfico de precios. Se calcula a partir de una media móvil (simple o exponencial) sobre el precio de cierre a la que envuelven dos bandas que se obtienen de añadir y sustraer al valor de la media K desviaciones estándar (habitualmente, K = 2). La distancia entre las curvas superior e inferior, igual a cuatro desviaciones estándar, es por lo tanto una medida de la volatilidad del precio del activo. De acuerdo con el análisis técnico, el que los precios sobrepasen las bandas indica que el mercado está sobrecomprado (si lo hacen por arriba) o sobrevendido (si lo hacen por abajo).

```
gcc['Close 30 day MA'] = gcc['Close'].rolling(window=30).mean()
gcc['Upper'] = gcc['Close 30 day MA'] + 2*(gcc['Close'].rolling(window=30).std())
gcc['Lower'] = gcc['Close 30 day MA'] - 2*(gcc['Close'].rolling(window=30).std())
gccplot = gcc[['Close','Close 30 day MA','Upper','Lower']].plot(figsize=(16,6));
gccplot.set_title('Grupo Cementero de Chihuahua, bandas de Bollinger al Cierre con media movil mensual');
```

Ahora, veremos el Retorno acumulado. 

#Retorno acumulado 

A diferencia de la fluctuacion por dia, el cual nos dice los cambios por día, este tipo de retornos acumula mediante la formula: 

$i_{i}=(1+r_{t})*i_{t-1}$

```
gcc['Retorno Acumulado']=(1+gcc['Retorno']).cumprod()
gmex['Retorno Acumulado']=(1+gmex['Retorno']).cumprod()
ich['Retorno Acumulado']=(1+ich['Retorno']).cumprod()
lamo['Retorno Acumulado']=(1+lamo['Retorno']).cumprod()
mfris['Retorno Acumulado']=(1+mfris['Retorno']).cumprod()

```

Ahora graficaremos, esto nos respondera de forma grafica, ¿Que activo nos brinda el mas alto (o el mas bajo) de los retornos/desempeños en caso de invertir una unidad monetaria en dicha accion (digase, $1)? 

```
gcc['Retorno Acumulado'].plot(label='GCC',figsize=(16,8))
gmex['Retorno Acumulado'].plot(label='GMEX')
ich['Retorno Acumulado'].plot(label='ICHB')
lamo['Retorno Acumulado'].plot(label='LAMO')
mfris['Retorno Acumulado'].plot(label='MFRIS')
plt.legend();
```

#**Portafolio**  

En esta parte se pretende armar un portafolio equitativo con las cinco empresas analizadas, es decir invirtiendo 20% en cada una. 

Algunos conceptos a considerar aparte del anterior es el de 'relación de nitidez' tal que en ingles es conocido como Sharp Ratio. 

Note: 

The formula for the Sharpe Ratio: 

#$S=(\frac{R_{p}-R_{f}}{\sigma_{p}})$

*   $R_{p}$ is Expected Portafolio return. 
*   $R_{f}$ is Risk-Free Return 
*   $\sigma_{p}$ is Portfolio Standard Deviation

```
import pandas as pd 
import quandl
```

```
start = pd.to_datetime('2020-01-01')
end = pd.to_datetime('2020-12-31')
```

```
gcc = web.DataReader('GCC.MX','yahoo',start,end)
gmex = web.DataReader('GMEXICOB.MX','yahoo',start,end)
ich = web.DataReader('ICHB.MX','yahoo',start,end)
lamo = web.DataReader('LAMOSA.MX','yahoo',start,end)
mfris = web.DataReader('MFRISCOA-1.MX','yahoo',start,end)
#nota: Adj Close 

```

```
for stock_df in (gcc,gmex,ich,lamo,mfris):
    stock_df['Retorno normado'] = stock_df['Adj Close'] / stock_df.iloc[0]['Adj Close']
```

Se realizara un portafolio equivalente, tal que cada empresa representara un 20% del capital invertido en el ramo. 

```
list(zip((gcc,gmex,ich,lamo,mfris),[.2,.2,.2,.2,.2]))
```

```
for stock_df, allo in zip((gcc,gmex,ich,lamo,mfris),[.2,.2,.2,.2,.2]):
    stock_df['Allocation'] = stock_df['Retorno normado']*allo
```

```
gcc.head()
```

```
for stock_df in (gcc,gmex,ich,lamo,mfris):
    stock_df['Position Values'] = stock_df['Allocation']*1000000
#1000000 MNX es la cantidad invertida en el portafolio 
```

```
gcc.head()
```

```
all_pos_vals = [gcc['Position Values'],gmex['Position Values'],ich['Position Values'],lamo['Position Values'],mfris['Position Values']]
portfolio_val = pd.concat(all_pos_vals,axis=1)
```

```
portfolio_val.head()
```

```
portfolio_val.columns = ['GCC Pos','GMEX Pos','ICHB Pos','LAMO Pos','MFRIS Pos']
portfolio_val
```

```
portfolio_val['Total Pos'] = portfolio_val.sum(axis=1)
```

```
portfolio_val
```

```
import matplotlib.pyplot as plt
%matplotlib inline 
```

```
portfolio_val['Total Pos'].plot(figsize=(10,8))
plt.title('Total Portfolio Value')
```

**Por lo que de invertir en este portafolio, se habria tenido un mal rato entre los meses 02 y 05, pero al final del año se habrian obtenido ganancias** 

```
portfolio_val.drop('Total Pos',axis=1).plot(figsize=(10,8))
```
